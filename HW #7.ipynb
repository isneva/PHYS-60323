```python
def f(x):
    return (x**4-2*x+1)

N = 10
a = 0
b = 2
h = (b-a)/N

s = f(a) + f(b)
for k in range(1,N,2):
    s += 4*f(a+k*h)

for k in range(2,N,2):
    s += 2*f(a+k*h)
    

print((1/3)*h*s)
```

```python
from matplotlib import pyplot as plt
from numpy import arange
import numpy as np


def C_v(x):
    return (((x**4)*np.exp(x)))/(((np.exp(x)-1)**2))

V = 10**-3
p = 6.022*(10**28)
k_b = 1.38*(10**-23)
D = 428
N = 1000
a = 0.0
TempStr = []
Cvstr = []

for T in range(5,500):    
    b = D/T
    h = (b-a)/N
    s = 0.5*C_v(b)
    for k in range(1,N):
        s += C_v(a+k*h)    
    TempStr.append(T)
    Cvstr.append(9*V*p*k_b*((T/428)**3)*h*s)
    
fig1 = plt.figure(figsize=(20,10))
ax1 = fig1.add_subplot(111)
ax1.scatter(TempStr,Cvstr,s=15,c='b',alpha=0.5)
xlabel("Temperature")
ylabel("Specific Heat")
```
